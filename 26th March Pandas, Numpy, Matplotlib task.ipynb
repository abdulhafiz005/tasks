```python
firmRelativeFrequency = firmFrequency/total * 100
```

```python
Frequency = [maleFrequency, femaleFrequency, firmFrequency, total]
RelativeFrequecy = [int(maleFrequency/total*100), int(femaleFrequency/total*100), int(firmFrequency/total*100), 100]
genderTable = pd.DataFrame({'Frequency': Frequency, 'Relative Frequency': RelativeFrequecy}, index=['Male', 'Female', 'Firms', 'Total'])
```

```python
genderTable
```

```python
forChart = genderTable.drop('Total', axis=0)
```

```python
forChart.plot(kind='pie', subplots=True)
```
